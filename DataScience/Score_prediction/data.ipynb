```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
import csv
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestRegressor
import joblib
```

```python
# 学習データの読み込み
# 学習データの読み込み
train_df=pd.read_csv("/Users/tasuke0630/making/DataScience/Score_prediction/data.csv")
train_df.head()
```

```python
# 元データをコピー
df=train_df.copy()
df=df.rename(columns={df.columns[0] :"a"})

#使わないカラムを削除する
df2=df.drop(columns=["a","Opponent","Date"])
df2.head()
```

```python
list_1=[]
list_2=[]
list_3=[]
list_4=[]
list_5=[]
list_6=[]
list_7=[]
list_8=[]
tmp=[]

df3=df2.drop(columns=["Goals","Assists","keyP","Drb","Fouled","Off","Disp","UnsTch"])

warnings.simplefilter("ignore")

list_1=df2.loc[:,"Goals"]
for i in range(len(list_1)):
    tmp=list_1[i]
    list_1[i]=tmp.replace("\t","")
for i in range(len(list_1)):
    tmp=list_1[i]
    list_1[i]=tmp.replace("-","0")  
    
    
list_2=df2.loc[:,"Assists"]
for i in range(len(list_2)):
    tmp=list_2[i]
    list_2[i]=tmp.replace("\t","")
for i in range(len(list_2)):
    tmp=list_2[i]
    list_2[i]=tmp.replace("-","0")
    
    
list_3=df2.loc[:,"keyP"]
for i in range(len(list_3)):
    tmp=list_3[i]
    list_3[i]=tmp.replace("\t","")
for i in range(len(list_3)):
    tmp=list_3[i]
    list_3[i]=tmp.replace("-","0")
    
    
list_4=df2.loc[:,"Drb"]
for i in range(len(list_4)):
    tmp=list_4[i]
    list_4[i]=tmp.replace("\t","")
for i in range(len(list_4)):
    tmp=list_4[i]
    list_4[i]=tmp.replace("-","0")
    
    
list_5=df2.loc[:,"Fouled"]
for i in range(len(list_5)):
    tmp=list_5[i]
    list_5[i]=tmp.replace("\t","")
for i in range(len(list_5)):
    tmp=list_5[i]
    list_5[i]=tmp.replace("-","0")
    
    
list_6=df2.loc[:,"Off"]
for i in range(len(list_6)):
    tmp=list_6[i]
    list_6[i]=tmp.replace("\t","")
for i in range(len(list_6)):
    tmp=list_6[i]
    list_6[i]=tmp.replace("-","0")
    
    
list_7=df2.loc[:,"Disp"]
for i in range(len(list_7)):
    tmp=list_7[i]
    list_7[i]=tmp.replace("\t","")
for i in range(len(list_7)):
    tmp=list_7[i]
    list_7[i]=tmp.replace("-","0")
    
    
list_8=df2.loc[:,"UnsTch"]
for i in range(len(list_8)):
    tmp=list_8[i]
    list_8[i]=tmp.replace("\t","")
for i in range(len(list_8)):
    tmp=list_8[i]
    list_8[i]=tmp.replace("-","0")
    
```

```python
x1=pd.DataFrame(list_1)
x2=pd.DataFrame(list_2)
x3=pd.DataFrame(list_3)
x4=pd.DataFrame(list_4)
x5=pd.DataFrame(list_5)
x6=pd.DataFrame(list_6)
x7=pd.DataFrame(list_7)
x8=pd.DataFrame(list_8)
```

```python
train=pd.concat([df3,x1,x2,x3,x4,x5,x6,x7,x8],axis=1)
train
```

```python
#カテゴリカル変数の数値変換
tmp_Position=pd.get_dummies(train["Position"],prefix="Position")
train=pd.concat([train,tmp_Position],axis=1).drop(columns="Position")
train
```

```python
#trainのデータ
train_x=train.drop(columns=["Goals"])
train_y=train["Goals"]
```

```python
clf=RandomForestRegressor(n_jobs=-1, random_state=1400)
clf.fit(train_x,train_y)
```
