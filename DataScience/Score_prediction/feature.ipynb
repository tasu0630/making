```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
import csv
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestRegressor
import joblib
```

```python
df=pd.read_csv("/Users/tasuke0630/making/DataScience/Score_prediction/graph_material.csv")
train_df=df.copy()
train_df=train_df.rename(columns={df.columns[0] :"a"})

#使わないカラムを削除する
df2=train_df.drop(columns=["a"])
df2.head()
```

```python
#trainのデータ
train_x=df2.drop(columns=["Goals"])
train_y=df2["Goals"]
```

```python
clf=RandomForestRegressor(n_jobs=-1, random_state=1400)
clf.fit(train_x,train_y)
```

```python
#特徴量の重要度
feature=clf.feature_importances_

#特徴量の重要度を上から順に出力する
x1=pd.DataFrame({"number": range(len(feature)),"feature": feature[:]})
x2=x1.sort_values("feature",ascending=False).reset_index()

#特徴量の高い順（列の要素を表す）
x3=x2.loc[:,"number"]
```

```python
#特徴量の名前
label=train_x.columns[0:]

#特徴量の高い順（名前順）
x4=x2

warnings.simplefilter("ignore")

for i in range(len(label)):
    x4["number"][i]=label[x3[i]]
    
x4=x4.drop(columns="index")
x4.rename(columns={x4.columns[0] :"name"})
```
