```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import linear_model
import joblib
```

```python
# 学習データの読み込み
train_df=pd.read_csv("/Users/tasuke0630/python/DataScience/kaggle/Taitanic/train.csv")

# テストデータの読み込み
test_df=pd.read_csv("/Users/tasuke0630/python/DataScience/kaggle/Taitanic/test.csv")

# 学習データとテストデータを連結する
df=pd.concat([train_df,test_df],ignore_index=True)
#nullのCSV化
df_null=df.isnull().sum()
df_null.to_csv("null.csv")

# 元データをコピー
df2=df.copy()
```

```python
#corr()メソッドで相関関係が出る
correlation=train_df.corr()
correlation
```

```python
#相関係数0.１以下切り捨て
df3=df2.drop(columns=["Age","SibSp","Parch","Parch"])
```

```python
#使わないカラムを削除する
df4=df3.drop(columns=["Name","Ticket","Cabin"])
```

```python
df4
```

```python
df4.isnull().sum()
#print(df4.Fare.mode())
#sns.countplot("Embarked",data=df4)
```

```python
# 欠損値の補完
df4.Fare=df2.Fare.fillna(8.05)
df4.Embarked=df2.Embarked.fillna("S")
```

```python
#カテゴリカル変数の数値変換
tmp_Sex=pd.get_dummies(df4["Sex"],prefix="Sex")
df4=pd.concat([df4,tmp_Sex],axis=1).drop(columns="Sex")

tmp_Embarked=pd.get_dummies(df4["Embarked"],prefix="Embarked")
df4=pd.concat([df4,tmp_Embarked],axis=1).drop(columns="Embarked")
df4
```

```python
# 学習データに分割した結果を変数trainに格納する
train=df4[~df4.Survived.isnull()]

# テストデータに分割した結果を変数trainに格納する
test=df4[df4.Survived.isnull()]

#trainのデータ
train_x=train.drop(columns=["Survived","PassengerId"])
train_y=train["Survived"]
```

```python
clf=linear_model.LogisticRegression(max_iter=200);
clf.fit(train_x,train_y);
```

```python
a=clf.coef_
b=clf.intercept_ 

print("回帰係数:", a) 
print("切片:", b) 
print("決定係数:",clf.score(train_x,train_y))
```

```python
# 学習結果の検証
X_test=test.drop(columns=["Survived","PassengerId"])
y_predict=clf.predict(X_test)

# 検証結果の表示
print("検証結果：", y_predict)
```

```python
#type(y_predict)
result=pd.DataFrame(y_predict)

#列名に名前をつける
result2=result.rename(columns={result.columns[0]: "Survived"})

#PassngerIdを合わせる
l=[]
l2=[]
t_id=test_df.loc[:,"PassengerId"]
for i in range(len(t_id)):
               tmp={
                   "PassengerId":t_id[i]
               }
               l.append(tmp);
               
l2=pd.DataFrame(l);
result3=pd.concat([l2,result2],axis=1)


#予測結果を整数に変換する
result3.Survived=result3.Survived.astype(int)
result3.PassengerId=result3.PassengerId.astype(int)
```

```python
# CSVの作成
result3.to_csv("predict3.csv",index=False)
```
