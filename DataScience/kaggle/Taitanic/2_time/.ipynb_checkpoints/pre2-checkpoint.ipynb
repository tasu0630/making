```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
```

```python
# 学習データの読み込み
train_df=pd.read_csv("/Users/tasuke0630/making/DataScience/kaggle/Taitanic/train.csv")

# テストデータの読み込み
test_df=pd.read_csv("/Users/tasuke0630/making/DataScience/kaggle/Taitanic/test.csv")

# 学習データとテストデータを連結する
df=pd.concat([train_df,test_df],ignore_index=True)
#nullのCSV化
df_null=df.isnull().sum()
df_null.to_csv("null.csv")

# 元データをコピー
df2=df.copy()
```

```python
#corr()メソッドで相関関係が出る
correlation=train_df.corr()

#ターゲットとの相関性が高い上位n個の数値特徴のヒートマップ
n=7
plt.figure(figsize=(32,10))
sns.heatmap(train_df[correlation.nlargest(n,"Survived").index].corr(),annot=True,fmt=".2f",square=True,cbar=False)
```

```python
df3=df2.drop(columns=["SibSp","Age","Pclass","Parch"])
```

```python
df3
```

```python
df4=df3.drop(columns=["Name","Ticket","Cabin"])
```

```python
# 欠損値の補完
df4.Fare=df2.Fare.fillna(8.05)
df4.Embarked=df2.Embarked.fillna("S")
```

```python
#カテゴリカル変数の数値変換
tmp_Sex=pd.get_dummies(df4["Sex"],prefix="Sex")
df4=pd.concat([df4,tmp_Sex],axis=1).drop(columns="Sex")

tmp_Embarked=pd.get_dummies(df4["Embarked"],prefix="Embarked")
df4=pd.concat([df4,tmp_Embarked],axis=1).drop(columns="Embarked")
```

```python
# 学習データに分割した結果を変数trainに格納する
train=df4[~df4.Survived.isnull()]

# テストデータに分割した結果を変数trainに格納する
test=df4[df4.Survived.isnull()]

# SalePriceを削除
test=test.drop(columns=["Survived"])

# 正解をy_trainに格納する
y_train=train.Survived

# 特徴量をX_trainに格納する
X_train=train.drop(columns=["Survived"])
```

```python
# 決定木モデルの準備(回帰)
model=tree.DecisionTreeClassifier()

# 決定木モデルの作成
model.fit(X_train,y_train)

# 作成した決定木モデルを使った予測をおこなう
y_pred=model.predict(test)
```

```python
# 予測結果の確認をする
y_pred
```

```python
# テストデータと予測結果の大きさを確認する
len(test),len(y_pred)

# 予測結果をテストデータに反映する
test["Survived"]=y_pred

# 提出用のデータマートを作成する
pred_df=test[["PassengerId","Survived"]].set_index("PassengerId")

# 予測結果を整数に変換する
pred_df.Survived=pred_df.Survived.astype(int)
print(pred_df.Survived.head())

# CSVの作成
pred_df.to_csv("predict2.csv",index_label=["PassengerId"])
```
