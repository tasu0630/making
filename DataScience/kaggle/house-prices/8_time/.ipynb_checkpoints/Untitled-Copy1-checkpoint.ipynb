```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
# 学習データの読み込み
train_df=pd.read_csv("/Users/tasuke0630/making/DataScience/kaggle/house-prices/train.csv",index_col = 'Id')

# テストデータの読み込み
test_df=pd.read_csv("/Users/tasuke0630/making/DataScience/kaggle/house-prices/test.csv",index_col = 'Id')
```

```python
# 学習データとテストデータを連結する
df=pd.concat([train_df,test_df],ignore_index=True)
#nullのCSV化
df_null=train_df.isnull().sum()
df_null.to_csv("df_null.csv")
```

```python
train_df.select_dtypes(exclude="object").hist(bins=10,figsize=(25,19))
plt.show()

plt.figure(figsize=(25,19))
sns.heatmap(train_df.corr("spearman"),annot=True)
plt.show()
```

```python
null=train_df.loc[:,train_df.isnull().sum()>500]
train_df=train_df.drop(null,axis=1)
test_df=test_df.drop(null,axis=1)

encode=LabelEncoder()
for i in train_df.select_dtypes(include="object"):
    train_df[i]=encode.fit_transform(train_df[i])
    test_df[i]=encode.fit_transform(test_df[i])
```

```python
#trainのデータ
train_x=train_df.drop(columns=["SalePrice"])
train_y=train_df["SalePrice"]
```

```python
clf=RandomForestRegressor(n_jobs=-1, random_state=2525)
pipeline=Pipeline(steps=[("impute",IterativeImputer(max_iter=4)),("model",clf)])
pipeline.fit(train_df.drop("SalePrice",axis=1),train_df["SalePrice"])
```

```python
y_predict=pipeline.predict(test_df)
```

```python
#type(y_predict)
result=pd.DataFrame(y_predict)

#列名に名前をつける
result2=result.rename(columns={result.columns[0]: "SalePrice"})
test_df2=test_df.reset_index()
test_df3=test_df2["Id"]
result3=pd.concat([test_df3,result2],axis=1)

#予測結果を整数に変換する
result3.SalePrice=result3.SalePrice.astype(int)
result3.Id=result3.Id.astype(int)
```

```python
# CSVの作成
result3.to_csv("predict8.csv",index=False)
```
