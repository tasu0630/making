```python
# 学習データに分割した結果を変数trainに格納する
train=df3[~df3.SalePrice.isnull()]

# テストデータに分割した結果を変数trainに格納する
test=df3[df3.SalePrice.isnull()]
```

```python
#trainのデータ
train_x=train.drop(columns=["SalePrice"])
train_y=train["SalePrice"]
```

```python
def making_model():
    model=Sequential()
    model.add(Dense(10,input_dim=392,activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(1,activation="relu"))
    model.compile(loss="mean_squared_error", optimizer="adam")
    return model


KerasR=KerasRegressor(build_fn=making_model,epochs=50,batch_size= 30)
KerasR._estimator_type ="Regressor"

KerasR.fit(train_x, train_y)

Y_pred_train=KerasR.predict(train_x)

X_test=test.drop(columns=["SalePrice"])
y_predict= KerasR.predict(X_test)

print("Stacking model score: %.3f" % KerasR.score(X_test, y_predict))
```

```python
#type(y_predict)
result=pd.DataFrame(y_predict)

#列名に名前をつける
result2=result.rename(columns={result.columns[0]: "SalePrice"})
test_df2=test_df.reset_index()
test_df3=test_df2["Id"]
result3=pd.concat([test_df3,result2],axis=1)

#予測結果を整数に変換する
result3.SalePrice=result3.SalePrice.astype(int)
result3.Id=result3.Id.astype(int)
```
